```python
from dataclasses import dataclass

@dataclass(frozen=True)
class Coordinate:
    lat: float
    lon: float

    def __str__(self):
        ns = 'N' if self.lat >= 0 else 'S'
        we = 'E' if self.lon >= 0 else 'W'
        return f'{abs(self.lat):.1f}°{ns}, {abs(self.lon):.1f}°{we}'
```

```python
import typing
class DemoNTClass(typing.NamedTuple):
    a:int
    b:float=1.1
    c='spam'
DemoNTClass.a
DemoNTClass.b
DemoNTClass.c
o=DemoNTClass(1)
print(o.a)
print(o.b)
print(o.c)
```

```python
o.a=2
```

```python
o.b=2
```

* 原因： NamedTuple 的核心特性是不可变性。一旦实例被创建，它的所有字段值都不能被修改。
* 机制： a 和 b 是你在类定义中声明的字段。在实例 nt 创建后，nt.a 和 nt.b 只是用于访问值的“只读”属性。当你尝试给它们重新赋值时，Python 会阻止这个操作，并抛出 can't set attribute 错误。

```python
o.c=2
```

* 原因： 这个错误和上面类似，但信息略有不同，因为它涉及到一个非字段的类属性。
* 机制： 在你的定义中，c='spam' 并没有类型注解，所以它不被认为是 NamedTuple 的一个字段，而是一个普通的类属性。这个类属性 c 的值是 'spam'。当你通过实例 nt 访问 nt.c 时，由于实例本身没有 c 属性，Python 会向上找到类属性 c。这个类属性是只读的，因此尝试修改它（nt.c = 1）也会失败。

```python
o.z=1
```

*   **原因：** 这是最关键的一个错误，它揭示了 `NamedTuple` 的底层实现。
*   **机制：**
    1.  **无 `__dict__`**：为了保持轻量级和内存效率，`NamedTuple` 的实例**没有 `__dict__` 属性**。在普通的 Python 类中，`__dict__` 是一个字典，用来存储实例的所有属性。当你给一个普通对象添加新属性（如 `obj.new_attr = value`）时，这个属性就会被存入 `__dict__` 中。
    2.  **字段固定**：`NamedTuple` 的字段在类定义时就已经完全确定了（在这个例子中只有 `a` 和 `b`）。
    3.  **报错解读**：当你尝试执行 `nt.z = 3` 时，Python 首先检查 `nt` 是否已有 `z` 属性，发现没有。然后，它会尝试在 `nt` 的 `__dict__` 中创建这个新属性。但由于 `NamedTuple` 实例根本没有 `__dict__`，Python 无法完成这个操作，于是抛出了这个非常明确的错误，告诉你：这个对象既没有 `z` 属性，也没有 `__dict__` 来让你创建新属性。
