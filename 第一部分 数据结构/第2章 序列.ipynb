```python
class LED:
    def __init__(self):
        self.brightness = None
        self.color = None

    def set_brightness(self, ident, intensity):
        self.brightness = (ident, intensity)

    def set_color(self, ident, red, green, blue):
        self.color = (ident, red, green, blue)


class Robot:
    def __init__(self):
        self.leds = {0: LED(), 1: LED()}
        self.beep_calls = []
        self.neck_angles = []

    def beep(self, times, frequency):
        self.beep_calls.append((times, frequency))

    def rotate_neck(self, angle):
        self.neck_angles.append(angle)

    def handle_command(self, message):
        match message:
            case ['BEEPER', frequency, times]:
                self.beep(times, frequency)
            case ['NECK', angle]:
                self.rotate_neck(angle)
            case ['LED', ident, intensity]:
                self.leds[ident].set_brightness(ident, intensity)
            case ['LED', ident, red, green, blue]:
                self.leds[ident].set_color(ident, red, green, blue)
            case _:
                raise ValueError(f'Unknown command: {message}')
```

```python
import unittest


class TestRobotHandleCommand(unittest.TestCase):
    def setUp(self):
        self.robot = Robot()

    def test_beeper_command(self):
        self.robot.handle_command(['BEEPER', 440, 3])
        self.assertEqual(self.robot.beep_calls, [(3, 440)])

    def test_neck_command(self):
        self.robot.handle_command(['NECK', 45])
        self.assertEqual(self.robot.neck_angles, [45])

    def test_led_brightness_command(self):
        self.robot.handle_command(['LED', 0, 80])
        self.assertEqual(self.robot.leds[0].brightness, (0, 80))

    def test_led_color_command(self):
        self.robot.handle_command(['LED', 1, 255, 128, 64])
        self.assertEqual(self.robot.leds[1].color, (1, 255, 128, 64))

    def test_unknown_command_raises(self):
        with self.assertRaises(ValueError):
            self.robot.handle_command(['MOTOR', 1])


suite = unittest.defaultTestLoader.loadTestsFromTestCase(TestRobotHandleCommand)
unittest.TextTestRunner(verbosity=2).run(suite)
```
